<span style="color:blue">D√©finition de la fonction Quantile $Q$.</span> Soit $f$ une densit√© par rapport √† la mesure de Lebesgue sur $\mathbb{R}$. $f$ est suppos√©e continue et strictement positive sur un intervalle $I$ (et nulle en dehors de $I$). 
La fonction de r√©partition associ√©e, not√©e $F$ est
$$
F :\mathbb{R} \to [0,1], \qquad \qquad  F(x) :=  \int_{-\infty}^x f(u) \mathrm{d} u, \qquad x \in \mathbb{R}.
$$

Les hypoth√®ses sur $f$ entrainent que $F$ est continue et strictement croissante sur $I$; on d√©finit la fonction  quantile $Q : [0,1] \to \mathbb{R} \cup \{\pm \infty\}$ par
$$
Q(0) := \inf I, \qquad  Q(u) := F^{-1}(u) \ \textrm{pour tout $u \in (0,1)$}, \qquad  Q(1):=\sup I.
$$


>Entre autre en finance, on consid√®re souvent la fonction $\operatorname{VaR}$ (pour *Value at Risk*) d√©finie par
$$
\operatorname{VaR}(\alpha) := Q(1-\alpha) \qquad
\quad  0<\alpha<1,
$$
>o√π les valeurs consid√©r√©es de $\alpha$ sont typiquement proches de z√©ro. Cette fonction est utilis√©e pour mesurer le risque : si  les pertes sont mod√©lis√©es commes des v.a. de densit√© $f$, $\operatorname{VaR}(0.01)$ repr√©sente la valeur des pertes qui n'est d√©pass√©e que dans $1 \%$ des cas.}



Le but de cette partie est d'estimer la fonction $Q$, √† partir de $n$ donn√©es $(x_1, \ldots, x_n)$ que l'on mod√©lise comme la r√©alisation de $(X_1, \ldots, X_n)$ o√π les v.a. $X_i$ sont i.i.d. de loi $f$. 
> Dans "la vraie vie", les mesures $x_1, \ldots, x_n$ peuvent √™tre obtenues √† l'aide de simulations (typiquement lorsque l'on veut calculer la fonction $\operatorname{VaR}$ pour une loi $f$ sp√©cifi√©e), ou √™tre des observations historiques (par exemple, lorsque l'on veut calculer la fonction $\operatorname{VaR}$ associ√©e √† une loi de perte dont on n'observe que des r√©alisations).

Etant donn√©es les v.a. $X_1, \cdots, X_n$, d√©finissons la fonction de r√©partition empirique sur $\mathbb{R}$ 
$$
F_n(x) := \frac 1 n \sum_{i=1}^n 1_{X_i\leq x}, \qquad  x \in \mathbb{R};
$$
et les statistiques d'ordre associ√©es $(X_{(i,n)})_{1\leq i\leq n}$ : 
$$
X_{(1,n)} \leq X_{(2,n)} \leq \cdots \leq X_{(i,n)} \leq \cdots \leq X_{(n,n)}.
$$
√Ä noter que ces in√©galit√©s sont strictes avec probabilit√© $1$ puisque les v.a. sont i.i.d. sous une loi √† densit√©. 


On consid√®rera successivement le cas o√π $f$  est la densit√© d'une loi
- **gaussienne standard**. *Indications : On pourra utiliser la fonction `randn` de `numpy.random` pour  les simuler, et les fonctions `norm.cdf` et `norm.ppf` de `scipy.stats` pour obtenir les valeurs exactes de $F$ et $Q$.  
- **loi gamma**. *Indications : On pourra utiliser les fonctions `numpy.random.gamma` et `scipy.stats.gamma` en choisissant leurs param√®tres `shape` et `scale`*.

<span style="color:blue">Question 1.1 (th√©orique).</span>   Pour tout $x \in \mathbb{R}$, exprimer $F_n(x)$ √† l'aide de $x$ et des statistiques d'ordre.

***     
__SOLUTION :__  Nous avons 
$$
F_n(x) := \frac 1 n \sum_{i=1}^n 1_{X_{i}\leq x}, \qquad  x \in \mathbb{R}, 
$$
qui est aussi √©gal √† 
$$
F_n(x) = \frac 1 n \sum_{i=1}^n 1_{X_{(i,n)}\leq x}, \qquad  x \in \mathbb{R}.
$$
Soit $x \in \mathbb{R}$; par d√©finition des $X_{(i,n)}$ il vient pour tout $ i= \{1, \cdots, n-1\}$
$$
F_n(x)= \left\{ \begin{array}{ll}
0 & \mbox{ si } x < X_{(1,n)}
\\
\frac i n & \mbox{ si } \quad X_{(i,n)} \leq x < X_{(i+1,n)}
\\
1 & \mbox{ si } x \ge X_{(n,n)}
\end{array} \right.
$$
***

<span style="color:blue">Question 1.2.</span> Le th√©or√®me de Glivenko-Cantelli affirme que presque-s√ªrement, $\lim_n \sup_{x \in \mathbb{R}} \big|F_n(x)-F(x) \big| =0$. Illustrer graphiquement ce r√©sultat dans les deux cas : l'√©chantillon $(X_1, \cdots, X_n)$ est i.i.d. de loi gaussienne standard, puis  il est i.i.d. de loi gamma.

```python
print("Question 1.2 Cas Gaussien")

## nombre de simulations
n = int(1e2)

## Gaussiennes centrees reduites
X = np.random.randn(n)

plt.figure(figsize=(15, 10))

#Fct de repartition theorique
x = np.linspace(min(X), max(X), 100)
F_x = sps.norm.cdf(x)
plt.plot(x, F_x, "r", linewidth=1.0, label="Fct de rep")
plt.title("Cas Gaussien, n=%1.0f" %n)



#Fct de repartition empirique 
X.sort()   # A comparer √† : Y = sorted(X)
F_x_n = np.arange(1, n+1) / n   #attention au dernier point

plt.step(X, F_x_n, "b", where="post", label="Fct de rep empirique") #comparer √† where="pre"


plt.axis([x[0]-0.5, x[-1]+0.5, 0, 1.1])

#On choisit la position de la legende
plt.legend(loc=4)
plt.show()
```

<span style="color:blue">D√©finition de la fonction Quantile empirique $Q_n$.</span>  √Ä la fonction de r√©partition empirique $F_n$, on associe la fonction _quantile empirique_ $Q_n$ d√©finie  sur $]0,1[$ par $$ Q_n(u) := \inf\{x\in \mathbb{R} : F_n(x) \geq u \}.$$

<span style="color:blue">Question 1.3. (th√©orique).</span> Soit $u \in (0,1)$. Montrer que $$Q_{n}(u) = X_{(\lceil nu \rceil,n)},$$ o√π $\lceil y\rceil$ d√©signe la partie enti√®re sup√©rieure d'un r√©el $y$.
> En python, la partie enti√®re sup√©rieure s'obtient par `numpy.ceil(y)`. 

***  
__SOLUTION__ : Soit $i \in \{0, \cdots, n-2\}$. ¬†
    D'apr√®s la Question 1.1, pour tout $u \in ]\frac i n,\frac{i+1}n]$ on a $$Q_n(u)=X_{(i+1,n)}.$$
    
Lorsque $i=n-1,$ pour tout $u \in ]\frac{n-1}{n},1]$,  on a $Q_n(u) = X_{(n,n)}$.

Ainsi,  dans tous les cas, nous avons $Q_n(u)=X_{(\lceil nu \rceil, n)}$  
***

<span style="color:blue">Question 1.4.</span> Montrer que  presque-s√ªrement, pour  tout $u \in ]0,1[$, $\lim_n Q_n(u) = Q(u)$.

Illustrer graphiquement cette convergence presque-s√ªre. On pourra comparer la suite des quantiles empiriques √† la valeur exacte $Q(u)$, calculable pour les choix de densit√©s $f$ consid√©r√©s ici √† l'aide des fonctions de `scipy.stats`.

***
__SOLUTION :__   Soit $u \in ]0,1[$. On √©crit
$$
|F(Q_n(u))-F(Q(u))| \le |F(Q_n(u))-F_n(Q_n(u))|+|F_n(Q_n(u))-F(Q(u))| \leq ||F-F_n||_{\infty} + |F_n(X_{(\lceil nu \rceil,n)})-F(Q(u))| .
$$
Par le th√©or√®me de Glivenko-Cantelli,  le premier terme tend vers zero  p.s. quand $n \to \infty$.   

Pour le second terme, nous avons par d√©finition de $F$ et de $F_n$ : 
$$
F_n(X_{(\lceil nu \rceil,n)}) = \frac{\lceil nu\rceil}{n} \qquad F(Q(u)) = u;
$$
dont nous d√©duisons qu'il tend aussi vers zero (de fa√ßon d√©terministe).

Ainsi, presque-s√ªrement, $\lim_n F(Q_n(u)) = F(Q(u))=u$ pour tout $u \in ]0,1[$. 

En utilisant la continuit√© de $F^{-1}$, nous en d√©duisons que presque s√ªrement, $\lim_n Q_n(u) = Q(u)$  pour tout $u \in ]0,1[$. 
***

```python
print("Question 1.4 cas Gaussien")

##nombre de simulations
n = 5000

#Vraie valeur du quantile 
u = 0.01
Q_u = sps.norm.ppf(u)


# R√©alisations al√©atoires
X = np.random.randn(n)

# Evolution du quantile empirique
Q_empirique_u = [ np.sort(X[0:i])[int(np.ceil(i*u))-1] for i in range(1, n+1) ] 


# Affichage: convergence du quantile empirique
plt.figure(figsize=(15, 10))
plt.axhline(y=Q_u, color="r", linewidth=1.5, label="Quantile Q(u)")
plt.plot(range(1,n+1), Q_empirique_u, color="b", linewidth=1.0, label="Quantile empirique")
plt.ylim(Q_u*0.5, Q_u*1.5)
plt.legend(loc="best")
plt.title("Cas Gaussien, u={}".format(u))
plt.show()
```

```python

print("Question 1.4 cas Gamma \n")

##Parametres de la loi Gamma
shape = 2
scale = 3
print(f"Parametres de la loi Gamma : shape={shape}, scale={scale}")

##nombre de simulations
n = 5000


#Vraie valeur du quantile 
u = 0.99
Q_u = sps.gamma.ppf(u, shape, scale=scale)
X = np.random.gamma(shape, scale=scale, size=n)

#Evolution du quantile empirique 
Q_empirique_u = [np.sort(X[0:i])[int(np.ceil(i*u))-1]
                 for i in range(1,n+1)] 

#Affichage: convergence du quantile empirique 
plt.figure(1)
plt.axhline(y=Q_u, color="r", linewidth=1.5, label="Quantile Q(u)")
plt.plot(range(1,n+1), Q_empirique_u, color="b", linewidth=1.0, label="Quantile empirique")
plt.ylim(Q_u*0.5, Q_u*1.5)
plt.title("Cas Gamma, u={}".format(u))
plt.legend(loc="best")
plt.show()
```

### <span style="color:blue">TCL pour l'estimateur $Q_n(u)$.</span> On peut d√©montrer (voir transparents de l'amphi) que pour tout $u \in \, ]0,1[$,

$$
\sqrt{n}\frac{f(Q(u))}{\sqrt{u(1-u)}}\left(Q_{n}(u)-Q(u)\right) \quad  \stackrel{\text{loi}}{\longrightarrow} \quad  \mathcal{N}(0,1). \qquad \qquad (1)
$$

Cela signifie que, lorsque $n$ est grand, les fluctuations de  $Q_n(u)$ autour de  $Q(u)$ se comportent en loi comme une gaussienne centr√©e, d'√©cart-type d'ordre $\frac{ \sqrt{u(1-u)}}{\sqrt{n} \, f(Q(u))}$.  Ce r√©sultat peut en particulier servir √† donner une estimation par intervalles de confiance de la quantit√© $Q(u)$ inconnue.  En pratique, n√©anmoins, on est confront√© √† deux difficult√©s dans l'√©valuation de l'√©cart-type de l'erreur normalis√©e :  
- le fait que $Q(u)$ n'est pas connu. Le lemme de Slutsky justifie n√©anmoins de remplacer $Q(u)$ par l'estimateur $Q_n(u)$.  
- le fait que la densit√© $f$ n'est pas connue.

<span style="color:blue">Question 1.5. Intervalle de confiance, $f$ connue</span>  Dans cette question, on suppose que l'expression de la densit√© $f$ est connue (mais $Q(u)$ n'est pas connue).

1. Soit $u \in ]0,1[$. D√©duire de la convergence en loi (1) l'expression d'un intervalle de confiance asymptotique de probabilit√© de couverture $95 \%$ pour la quantit√© $Q(u)$. 

2. Dans le cas o√π $f$ est la densit√© d'une loi gaussienne standard, puis le cas o√π c'est une loi gamma : calculer cet intervalle de confiance √† partir de $n$ r√©alisations de v.a. $X_1, \cdots, X_n$ de loi $f$, et illustrer la convergence en loi (1).


***
#### __Calculs pour l'intervalle de confiance :__ 

A partir de la convergence en loi (1),
$$ \lim_n \mathbb{P}\bigg( Q(u) \in \mathcal{I}_n^\star \bigg) = 0.95 $$
en ayant d√©fini l'intervalle al√©atoire
$$
\mathcal{I}_n^\star := \left[ Q_n(u)-1.96 \frac{\sqrt{u(1-u)}}{\sqrt{n} \, f(Q(u))}; Q_n(u)+1.96 \frac{\sqrt{u(1-u)}}{\sqrt{n} \, f(Q(u))}  \right] 
$$

Cet intervalle d√©pend de la quantit√© inconnue $Q(u)$. Il est donc inexploitable. N√©anmoins, nous disposons de $Q_n(u)$, un estimateur consistant de la quantit√© inconnue $Q(u)$; et nous savons que $f$ est continue. Donc nous avons $\lim_n f(Q_n(u)) = f(Q(u))$ (p.s. donc en probabilit√©s). Par le lemme de Slutsky,  nous avons aussi
$$
\lim_n \mathbb{P}\bigg( Q(u) \in \mathcal{I}_n\bigg) = 0.95
$$
en ayant pos√©
$$
\mathcal{I}_n := \left[ Q_n(u)-1.96 \frac{\sqrt{u(1-u)}}{\sqrt{n} \, f(Q_n(u))}; Q_n(u)+1.96 \frac{\sqrt{u(1-u)}}{\sqrt{n} \, f(Q_n(u))}  \right].
$$
***

```python
print("**Question 1.5 cas Gaussien** \n")

##nombre de simulations
n = int( 1e4)

#Vraie valeur du quantile 
u = 0.99
Q_u = sps.norm.ppf(u)


## R√©alisations al√©atoires
X = np.random.randn(n)

## calcul et affichage de l'IC asymptotique
Q_n_u = np.sort(X)[int(np.ceil(u*n)) - 1]
rayonIC = 1.96 * np.sqrt(u*(1-u)) / (np.sqrt(n)* sps.norm.pdf(Q_n_u))
IC = np.array([Q_n_u-rayonIC, Q_n_u+rayonIC])

print("Vrai quantile au niveau 1-alpha={:1.4f}: {:1.2f} \n".format(u, Q_u))
# l'instruction 1.2f permet de choisir le nombre de chiffres significatives

print("Valeur estimee : {:1.2f},\n IC= [{:1.2f}, {:1.2f}] \n".format(Q_n_u, IC[0], IC[1]))

flag = np.logical_and(Q_u < IC[1], Q_u>IC[0])
print("Vraie valeur dans l'intervalle: {} \n".format(flag))




## Histogramme de l'erreur  
###########################
# nombre de repetitions (pour l'histogramme)
m = int(1e4)

X = np.random.randn(m, n)

plt.figure(2)

# echantillon des quantiles empiriques Q_n(u)
X.sort(axis=1)
Q_n_u = np.array(X[:, int(np.ceil(n*u))-1]) # echantillon de taille m
erreur = Q_n_u - Q_u

minE, maxE = min(erreur), max(erreur)

# choix du nombre de colonnes (voir Exo 4)
N = m**(1./3)*(maxE - minE) / (3.5*np.std(erreur))

# Affichage de l'histogramme de l'erreur 
plt.hist(erreur, bins=int(N), density="true",label="histogramme")
plt.title("Histogramme d'erreurs Q_n(u)-Q(u) pour u={}".format(u))

# Densite gaussienne de variance connue, pour comparaison
s = np.sqrt(u*(1.-u)) / (sps.norm.pdf(Q_u) * np.sqrt(n))
x = np.linspace(minE, maxE, 100)
gaussienneLimite = sps.norm.pdf(x, loc=0, scale=s)
plt.plot(x, gaussienneLimite, "r", linewidth=1.5, label="Densite gaussienne de variance $u(1-u)/f(Q(u))^2 n$")

# pour faire apparaitre la legende en dehors du graphique
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)
plt.show()



```

```python

print("**Question 1.5 cas Gamma** \n")

# parametres de la loi Gamma
shape = 2
scale = 3
print("Parametres de la loi Gamma : shape={}, scale={}\n".format(shape,scale))

# nombre de simulations
n = 20000

# vraie valeur du quantile 
u = 0.99
Q_u = sps.gamma.ppf(u, shape, scale=scale)

# calcul du quantile empirique √† l'aide de n r√©alisations iid de la loi
X = np.random.gamma(shape, scale=scale, size=n)
Q_n_u = np.sort(X)[int(np.ceil(n*u))-1]

# calcul et affichage de l'IC asymptotique
rayonIC = 1.96*np.sqrt(u*(1-u))/(np.sqrt(n)* sps.gamma.pdf(Q_n_u, a=shape, scale=scale))
IC = np.array([Q_n_u-rayonIC, Q_n_u+rayonIC])

print("Vrai quantile d'ordre {:1.4f}: {:1.2f} \n".format(u, Q_u))
print("Valeur estimee et IC au niveau 1-alpha: {:1.2f},\n IC= [{:1.2f}, {:1.2f}] \n".format(Q_n_u,IC[0],IC[1]))

flag = np.logical_and(Q_u < IC[1], Q_u>IC[0])
print("Vraie valeur dans l'intervalle: {} \n".format(flag))


# Histogramme de l'erreur  
##########################
# nombre de repetitions (pour l'histogramme)
m = 1000

X = np.random.gamma(shape, scale=scale, size=(m,n))

plt.figure(2)
plt.close()

# echantillon des quantiles empiriques Q_n(u)
X.sort(axis=1)
Q_n_u = np.array(X[:, int(np.ceil(n*u))-1]) 
erreur = Q_n_u - Q_u

minE, maxE = min(erreur), max(erreur)

##choix du nombre de colonnes
N = m**(1./3)*(maxE-minE) / (3.5*np.std(erreur))

plt.hist(erreur, bins='auto', density="true",label="Erreur")
plt.title("Histogramme d'erreurs Q_n(u)-Q(u), pour u ={}".format(u))

# densite gaussienne de variance connue pour comparaison
s = np.sqrt(u*(1.-u)) / (sps.gamma.pdf(Q_u, a=shape, scale=scale) * np.sqrt(n))
print("Valeur de l ecart-type de la loi limite: {} \n".format(s))

x = np.linspace(minE, maxE, 100)
gaussienneLimite = sps.norm.pdf(x, loc=0, scale=s)
plt.plot(x,gaussienneLimite,"r", linewidth=1.5, label="Densite loi Gamma de variance connue $u(1-u)/f(Q(u))^2 n$")

# pour faire apparaitre la legende en dehors du graphique
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)
plt.show()
```

***
$\newcommand{\Var}{\operatorname{Var}}$
<span style="color:blue">Question 1.6. Intervalles de confiance, $f$ inconnue.</span> 
Dans les cas o√π $f$ n'est pas connue, on peut proc√©der √† une estimation de la variance asymptotique de l'erreur renormalis√©e apparaissant dans le TCL (voir (1)) afin de pouvoir fournir un intervalle de confiance. Cela est possible, quitte √† simuler plusieurs fois l'estimateur $Q_n(u)$.

Nous allons voir une technique alternative, qui utilise un seul jeu de $n$ simulations (le m√™me que celui utilis√© pour le calcul de $Q_n(u)$).

1. Montrer que  pour tout $u \in ]0,1[$, 
$$ \sqrt{n} (F_n(Q(u))- F(Q(u)))  = \sqrt{n} (F_n(Q(u))- u)  \stackrel{loi}{\longrightarrow}  \mathcal{N}(0,u(1-u)).$$


2. Soit $u \in ]0,1[$; on pose
$$  u^-_n := u  - 1.96  \,  \frac{\sqrt{u (1-u)}}{\sqrt{n}}, \qquad u^+_n := u  + 1.96 \,  \frac{\sqrt{u (1-u)}}{\sqrt{n}}.$$ 
Montrer que $[Q_n(u_n^-),Q_n(u_n^+)[$ est un intervalle de confiance asymptotique de probabilit√© de couverture  $95\%$ pour $Q(u)$.


3. Illustrer num√©riquement ce calcul d'intervalle de confiance pour des valeurs de $u=1-\alpha$ allant de $\alpha=10^{-2}$ √† $\alpha=10^{-5}$.

***
__SOLUTION__ 

1. Par d√©finition de $F_n$, nous avons
$$ 
F_n(Q(u)) = \frac{1}{n} \sum_{i=1}^n 1_{X_i \leq Q(u)}
$$
est la moyenne empirique de $n$ variables al√©atoires i.i.d. de Bernoulli de param√®tre de succ√®s 
$$ \mathbb{P}(X \leq Q(u)).
$$ 
Puisque par hypoth√®se $F$ est continue, il vient
$$
\mathbb{P}(X \leq Q(u)) = F(Q(u))=u.
$$ 
Le r√©sultat est alors une cons√©quence du TCL classique pour des v.a. de Bernoulli de param√®tre de succ√®s $u$.
*** 

***
2. D'apr√®s la question pr√©c√©dente,,
$$ \lim_n \mathbb{P}\bigg(u_n^- \leq F_n(Q(u)) < u_n^+ \bigg) =  0.95.$$

Or, toute fonction de r√©partition $\tilde F$, d'inverse g√©n√©ralis√©e $\tilde F^{-1}$, v√©rifie la relation :
$$
\tilde F(x) \geq u \Longleftrightarrow x \geq \tilde F^{-1}(u).$$
On en d√©duit que 
$$ \lim_n \mathbb{P}(Q(u) \in [Q_n(u_n^-),Q_n(u_n^+)[)  = 0.95.$$
***

```python
print("*** Question 1.6 Cas Gaussien *** \n")
alpha = 1.e-2

# nombre de simulations de X
n = int(5e4)

X = np.random.randn(n)
X.sort()

# Vraie valeur de la VaR (ici on voit u=1-alpha)
Q_u = sps.norm.ppf(1-alpha)


# quantiles empiriques
unmoins = 1-alpha - 1.96*np.sqrt(alpha*(1-alpha)/n)
unplus = min(1,1-alpha + 1.96*np.sqrt(alpha*(1-alpha)/n))


Q_n_u = X[int(np.ceil(n*(1-alpha)))-1]

index_moins = int(np.ceil(n*unmoins))-1
index_plus = int(np.ceil(n*unplus))-1


IC = X[[index_moins, index_plus]]

print("Vrai quantile au niveau 1-alpha={:1.5f}: {:1.6f} \n".format(1-alpha, Q_u))
##l'instruction 1.2f permet de choisir le nombre de chiffres significatives
print("Bornes u- et u+ :[{:1.6f}, {:1.6f}] \n".format(unmoins,unplus))
print("Valeur estimee du quantile: {:1.6f}, \n  IC= [{:1.6f}, {:1.6f}] \n".format(Q_n_u, IC[0], IC[1]))

flag = np.logical_and(Q_u < IC[1], Q_u>IC[0])
print("Vraie valeur dans l'intervalle: {} \n".format(flag))

```

<span style="color:magenta"> **V√©rification de l'IC**.   
    Pour plusieurs valeurs de la taille de l'√©chantillon $n$, on peut √©galement simuler 100 r√©alisations ind√©pendantes de cet intervalle de confiance puis afficher l'√©volution, en fonction de $n$, de la proportion d'IC contenant la vraie valeur $Q(u)$.  </span> 
    

```python
alpha = 1.e-3
u=1-alpha

print("*** Cas Gaussien - qualite IC *** \n")
Q_u = sps.norm.ppf(1-alpha)

## nombre max de simulations de X
n = int(1e5)
## nombre de realisations pour l'IC
m = 100
## R√©alisations al√©atoires 
X = np.random.randn(m, n)



## Etant donn√© une taille d'√©chantillon n, et un tableau de r√©alisations ar,
## calcul de m IC, chacun √† l'aide de n variables
## et compte la proportion d'IC qui contiennent le vrai quantile
def ProportionDansIC(ar, n):
    arS = np.sort(ar[:, :(n+1)], axis=1)
    unmoins = u-1.96*np.sqrt(u*(1-u)/n)
    unplus = min(1,u+1.96*np.sqrt(u*(1-u)/n))
    return np.sum( np.logical_and(Q_u < arS[:, int(np.ceil(n*unplus))-1], Q_u>arS[:, int(np.ceil(n*unmoins))-1]) ) / m


## Etant donn√©e une taille d'√©chantillon n, et un tableau de r√©alisations ar,
## calcul de m IC, chacun √† l'aide de n variables
## et calcule la valeur moyenne de la taille relative des IC (par rapport au quantile empirique)
def MoyErreurRelative(ar, n):
    arS = np.sort(ar[:, :(n+1)], axis=1)
    unmoins = u-1.96*np.sqrt(u*(1-u)/n)
    unplus = min(1,u+1.96*np.sqrt(u*(1-u)/n))
    lowerbound = arS[:, int(np.ceil(n*unmoins))-1]
    upperbound = arS[:, int(np.ceil(n*unplus))-1]
    estim = arS[:, int(np.ceil(n*u))-1]
    Delta = (upperbound-lowerbound)/estim
    return np.mean(Delta)



## Affichage
############
plt.figure(1)
# diff√©rence entre deux tailles d'√©chantillon
pas = int(n/10)
# proportion d'IC 
plt.scatter(np.arange(100,n+1,pas),
            [ProportionDansIC(X, i) for i in range(100, n+1, pas)],
            color="b", linewidth=1.0,
            label="Proportion d'IC contenant le quantile pour 1-alpha={}".format(1-alpha))
plt.axhline(y=0.95, color="k")
plt.legend(loc="best")
plt.show()

plt.figure(2)
# diff√©rence entre deux tailles d'√©chantillon
pas = int(n/10)
plt.scatter(np.arange(100,n+1,pas),
            [MoyErreurRelative(X, i) for i in range(100, n+1, pas)],
            color="b", linewidth=1.0,
            label="Valeur moy de la precision relative")
plt.legend(loc="best")
plt.show()
```

<span style="color:blue">Esp√©rance conditionnelle de queue.</span>  Une fois le quantile estim√©, il est parfois int√©ressant d'estimer l'*Expected Shortfall*, qui se d√©finit comme √©tant l'esp√©rance conditionnelle de queue $$E_{1-u}:= \mathbb{E}[X_1\,|\, X_1> Q(u)], \qquad X_1 \sim f. $$ 

<span style="color:blue">Question 1.7.</span>  Dans le cas o√π $f$ est la loi gaussienne centr√©e r√©duite, puis dans le cas o√π $f$ est la loi gamma : donner un intervalle de confiance asymptotique de probabilit√© de couverture  $95\%$ pour l'esp√©rance conditionnelle de queue et l'illustrer num√©riquement.

> Dans cette question, on pourra se contenter d'estimer $Q(u)$ par le quantile empirique, et ensuite utiliser le fait que les variables al√©atoires $\{X_i: X_i > X_{(\lceil nu\rceil,n)} \}$ sont approximativement i.i.d. et distribu√©es selon la loi de $X$ sachant $X> Q(u)$. Plus pr√©cis√©ment, on pourra utiliser le fait que les variables al√©atoires $\{X_i: X_i > X_{(\lceil n u\rceil,n)} \}$ satisfont √† un th√©or√®me central limite: le Th√©or√®me 4.1 (iii) page 80 [dans ce document](https://tel.archives-ouvertes.fr/tel-00743159/document) fournit des √©l√©ments de th√©orie.

Dans le cas de la loi gaussienne centr√©e r√©duite, on pourra comparer avec la valeur exacte $ \quad  (1-u)^{-1} \, f(\Phi^{-1}(u)) \quad $
o√π $\Phi$ d√©signe la fonction de r√©partition de la loi gaussienne centr√©e r√©duite. 



```python
print("*** Question 1.7 : cas Gaussien*** \n")

## nombre de simulations
n = int(1e4)

## R√©alisations al√©atoires, et tri
X = np.random.randn(n)
X.sort()

## Quantile empirique 
alpha = 1.e-3
u = 1-alpha
Q_n_u = X[int(np.ceil(n*u))-1]


## S√©lection des r√©alisations v√©rifiant la condition 
# Methode 1
echantillonExpectedSF = X[ X > Q_n_u ]
# Methode plus rapide 
echantillonExpectedSF = X[int(np.ceil(n*u)):]
# Taille de l'√©chantillon 
tailleEchantillonES = echantillonExpectedSF.size


## Estimation de E_alpha
moyenneES = np.mean(echantillonExpectedSF)

s = np.std(echantillonExpectedSF) 
rayonIC = 1.96 * s / np.sqrt(tailleEchantillonES)

## Vraie valeur de E_alpha
EStheorique= sps.norm.pdf(sps.norm.ppf(u))/(1-u)


## Affichage 
print ("Vrai Expected ShortFall d'ordre u ={:1.3f}: {:1.3f} \n".format(u, EStheorique))

print("Taille echantillon pour le calcul de l'ES: {} \n".format(tailleEchantillonES))

print ("Valeur estimee  {:1.3f}  \n et erreur relative  {:1.3f} \n" .format(moyenneES,2*rayonIC/moyenneES))

```

```python
print ("*** Question 1.7 : cas Gamma*** \n")

## parametres de la loi Gamma
shape = 2
scale = 3
print("Parametres de la loi Gamma : shape={}, scale={}\n".format(shape,scale))

##  nombre de simulations
n = int(1e4)

## R√©alisations al√©atoires, et tri
X = np.random.gamma(shape, scale=scale, size=n)
X.sort()

## Quantile empirique 
alpha = 1.e-3
u = 1-alpha
Q_n_u = X[int(np.ceil(n*u))-1]


## S√©lection des r√©alisations v√©rifiant la condition 
echantillonExpectedSF = X[X > Q_n_u]

# Taille de l'√©chantillon 
tailleEchantillonES = echantillonExpectedSF.size

## Estimation de E_alpha 
moyenneES = np.mean(echantillonExpectedSF)

s = np.std(echantillonExpectedSF)
rayonIC = 1.96*s/np.sqrt(tailleEchantillonES)

print("Taille echantillon pour le calcul de l'ES: {} \n".format(tailleEchantillonES))

print ("Valeur estimee {:1.2f}  \n et erreur relative  {:1.2f} \n" .format(moyenneES,2*rayonIC/moyenneES))
```

__Calculs th√©oriques :__

Les tirages $X_i$ avec $X_i > X_{(\lceil n u \rceil,n)}$ sont distribu√©s selon la loi $\mathcal{L}(X_1| \{X_1 > X_{(\lceil n u \rceil, n)} \})$. Ainsi, d'apr√®s le th√©or√®me central limite √©voqu√©, si $Y_1, \ldots,Y_{N_n}$ d√©signent les variables al√©atoires $\{X_i: X_i > X_{(\lceil nu\rceil,n)} \}$, on a 

$$\sqrt{\frac{N_n}{\textrm{Var}(Y_1)}} \left( \frac{Y_1+ \cdots+Y_{N_n}}{N_n} - E_{1-u} \right) \rightarrow \mathcal{N}(0,1).$$

On notera que $N_n$ est ici d√©terministe vue la fa√ßon dont on a construit les v.a. $Y_i$ : on a $N_n = n-\lceil n u \rceil$ puisque $Y_1 = X_{(\lceil n u \rceil+1,n)}, Y_2 = X_{(\lceil n u \rceil+2,n)}, \ldots, Y_{N_n} = X_{(n,n)}$.

La quantit√© $\textrm{Var}(Y_1)$ √©tant inconnue, on l'estime par la variance empirique $\hat{\sigma}^2_n$ construite √† partir de   $Y_1, \ldots,Y_{N_n}$. Ainsi, en notant $\hat{m}_n := (Y_1+ \cdots+Y_{N_n})/N_n$,

$$
\lim_n \mathbb{P}\bigg( E_{1-u} \in \mathcal{I}_n \bigg) =  0.95,$$
en ayant pos√©
$$
\mathcal{I}_n := \left[ \hat{m}_n-1.96 \frac{\hat{\sigma}_n}{\sqrt{N_n}};\hat{m}_n+1.96 \frac{\hat{\sigma}_n}{\sqrt{N_n}} \right] 
$$

## Exercice 2. Estimation param√©trique
***

Dans cet exercice, on dispose de  donn√©es $x_{1}, \ldots,x_{n}$ mod√©lis√©es comme la r√©alisation d'un vecteur al√©atoire $(X_{1}, \ldots,X_{n})$ que l'on suppose √™tre un $n$-√©chantillon d'une loi _inconnue_.  On d√©finit un mod√®le statistique param√©trique : on va chercher la loi qui, dans une famille de lois param√©trique donn√©e $\{f_\theta, \theta \in \mathbb{R}^d \}$, s'ajuste au mieux aux donn√©es. Dans la suite, on prend pour famille de lois:

$$(\mathbb{R}, \mathcal{B}(\mathbb{R}), \{ \mathcal{N}(0,\sigma^2), \sigma^2 > 0 \}). $$

On rappelle que la variance d'une loi $\mathcal{N}(0, \sigma^2)$ est $\sigma^{2}$ et que son moment d'ordre $4$ vaut $3 \sigma^4$. 

<span style="color:blue">Question 2.1. Intervalles de confiance pour $\sigma$.</span> Nous allons d'estimer $\sigma$ √† partir de l'observation de $X_1, \ldots, X_n$. 

a)   Utiliser la statistique $M_n := n^{-1} \sum_{i=1}^n X_i^2$. Sous $\mathbb{P}_\sigma$, quelle est la loi de $n M_n/\sigma^2$ ? En d√©duire un intervalle de confiance non asympototique de probabilit√© de couverture $95 \%$, pour estimer $\sigma$. 
> Indication : on pourra utiliser la loi du $\chi^{2}$ et la fonction `sps.chi2.ppf`).

Application num√©rique

__Calculs th√©oriques__:  Sous $\mathbb{P}_\sigma$, $X_i^2/\sigma^2 \sim \chi^2(1)$, la loi du Chi-2 √† un degr√© de libert√©.  On en d√©duit que 
pour toute valeur de $n$,  $$\frac{n M_n}{\sigma^2} \sim \chi^2(n).$$
En notant $q^{\chi}_{n,\alpha}$ le quantile de niveau $\alpha$ de la loi $\chi^2(n)$, on a pour tout $n \ge 1$
$$
\begin{aligned}
1-\alpha = \mathbb{P}\Bigl(q^\chi_{n,\alpha/2} \le \frac{n M_n}{\sigma^2} \le q^\chi_{n,1-\alpha/2}\Bigr)
&= \mathbb{P}\Bigl(\frac{n M_n^2}{q^\chi_{n,1-\alpha/2}} \le \sigma^2 \le \frac{n M_n}{q^\chi_{n,\alpha/2}} \Bigr)
\\
&= \mathbb{P}\left(\sqrt{\frac{n M_n}{q^\chi_{n,1-\alpha/2}}} \le \sigma \le \sqrt{\frac{n M_n}{q^\chi_{n,\alpha/2}}} \right)
\end{aligned}
$$
$\left[\sqrt{n M_n/q^\chi_{n,1-\alpha/2}},\sqrt{n M_n/q^\chi_{n,\alpha/2}}\right]$ est donc un intervalle de confiance de niveau exactement √©gal √† $1-\alpha$ pour le param√®tre $\sigma$, pour tout $n$ (on choisira donc $\alpha = 0.05$).

```python
print ("*** Question 2.1 *** \n")

n = 1000
sigma = 2.
alpha = 0.05

## Simulation des r√©alisations gaussiennes variance sigma2
X = np.random.randn(n)
X = sigma*X

## Calcul de la statistique n*M_n
xSquare = X*X
sumSquare = np.sum(xSquare)




## IC non asymptotique
#######################

# Quantiles de la loi chi^2(n) de niveau 0.975 et 0.025
quantileHaut = sps.chi2.ppf(q = 1-alpha/2, df=n)
quantileBas = sps.chi2.ppf(q = alpha/2, df=n)

# Calcul de l'intervalle
ICnonAsymptotique = (np.sqrt(sumSquare/quantileHaut), np.sqrt(sumSquare/quantileBas))

# Affichage 
print("Vraie valeur de sigma : {} \n".format(sigma))
print("Taille de l'echantillon: {} \n".format(n))
print("Intervalle de confiance non asymptotique [{:1.3f}, {:1.3f}] \n".format(ICnonAsymptotique[0],ICnonAsymptotique[1]))
# Erreur relative, en 'trichant' i.e. en exploitant la vraie valeur de sigma
print("Erreur relative (triche ...): {:1.3f} \n".format((ICnonAsymptotique[1]-ICnonAsymptotique[0])/sigma))
hatsigma = np.sqrt(sumSquare/n)
print("Erreur relative (plus honn√™te): {:1.3f} \n".format((ICnonAsymptotique[1]-ICnonAsymptotique[0])/hatsigma))

```

b) D√©montrer que  $\sqrt{n} \left(  \sqrt{M_n }-\sigma \right)  \stackrel{loi}{ \rightarrow}  \mathcal{N}(0,\sigma^{2}/2)$.

***
__SOLUTION :__ 
Le TCL pour des v.a. i.i.d. ayant un moment d'ordre $2$ s'applique pour $M_n$; la variance asymptotique  est $$
\mathrm{Var}_\sigma(X_1^2)=\mathbb{E}_\sigma[X_1^4]-\mathbb{E}_\sigma[X_1^2]^2 = 3\sigma^4-\sigma^4 = 2\sigma^4.$$


Par la m√©thode delta (voir polycopi√© de MAP433 Th√©or√®me IV-5-47), le TCL pour l'estimateur $T(M_n) := \sqrt{M_n}$ s'√©crit : sous $\mathbb{P}_\sigma$,
$$
 \sqrt{n}\Bigl(\sqrt{M_n} - \sigma\Bigr) \stackrel{loi}{ \rightarrow} 
\mathcal{N}\left(0,(T'(\sigma^2))^2 Var(X_1^2)\right)
= \mathcal{N}\left(0, \sigma^2/2\right).
$$


Par la LGN pour des variables i.i.d. ayant un moment d'ordre $1$,  sous $\mathbb{P}_\sigma$ on a $\lim_n M_n = \mathbb{E}_\sigma[X^2] = \sigma^2$ presque-s√ªrement,  d'o√π $\lim_n \sqrt{M_n} = \sigma$ p.s. Le lemme de Slutsky assure donc que un intervalle de confiance asymptotique √† $95\%$ pour $\sigma$ est donn√© par 
$\left[\sqrt{M_n} \pm 1.96  \frac{\sqrt{M_n}}{\sqrt{2n}}\right]
= \sqrt{M_n} \left[1 \pm 1.96  \frac{1}{\sqrt{2n}}\right]$.

```python
## IC asymptotique
#####################
meanSquare = sumSquare/n
estimSigma = np.sqrt(meanSquare)


rayonIC = 1.96 * estimSigma / np.sqrt(2*n)

print("Valeur estimee par moyenne empirique: {:1.3f} \n".format(estimSigma))
print("Intervalle de confiance asymptotique [{:1.3f}, {:1.3f}]".format(estimSigma-rayonIC, 
      estimSigma+rayonIC))
print("Erreur relative {:1.3f}".format(2*rayonIC/estimSigma))
 
```

#### A. Th√©or√®me ergodique pour le processus $\operatorname{AR}(1)$ stationnaire ####

Un processus $\operatorname{AR}(1)$ gaussien stationnaire est d√©fini par r√©currence comme suit (pour simplifier, on se place en dimension $1$).  On fixe $\rho\in ]-1,1[$ et on pose pour $i\ge 1$,

$$
X_i := \rho X_{i-1}+\sqrt{1-\rho^2}  Y_i, \qquad \qquad  X_0 = Y_0 \qquad (3)
$$

o√π $(Y_i)_{i\ge 0}$ est une suite de variables al√©atoires i.i.d. de loi 
$\mathcal{N}(0, 1)$, ind√©pendante de $X_0$.
C'est donc un processus qui se renouvelle d'un instant √† l'autre en gardant une part de son ancienne valeur ($\rho X_{i-1}$) et en ajoutant une part d'_innovation_ $\sqrt{1-\rho^2} Y_i$.

De fa√ßon √©quivalente, nous avons 
$$
X_i\;=\;\rho^i X_0+\sqrt{1-\rho^2}\sum_{j=1}^{i}\rho^{i-j}Y_{j}. \qquad \qquad \qquad (4)
$$

> L'expression de $X_i$ dans l'√©quation (3) se pr√™te naturellement √† une impl√©mentation avec une boucle `for`.  Cette expression exprime clairement la propri√©t√© markovienne du processus $(X_i)_i$.  
L'√©quation (4) permet d'effectuer des calculs plus rapides que (3) en utilisant la programmation matricielle de `numpy` plut√¥t que des boucles, mais, d'un autre cot√©, elle peut mener √† des probl√®mes de pr√©cision lorsque $i\gg 1$, puisqu'elle am√®ne √† calculer des expressions du type $\rho^{\textrm{(tr√®s grand nombre)}}$.
Nous d√©conseillons donc l'usage de l'expression (4) en pratique. Sous l'angle de vue probabiliste, l'avantage de l'equation (4) est de permettre un calcul rapide de la loi de $X_i$ en tant que combinaison lin√©aire de v.a. gaussiennnes i.i.d. On peut v√©rifier que pour tout $i \geq 0$, $X_i \sim  \pi_\star := \mathcal{N}(0,1)$.

<span style="color:blue">Th√©or√®me ergodique.</span>  Il affirme que 
pour tout  $B\subset \mathbb{R}$ mesurable, presque s√ªrement, 
$$\frac{1}{n}\sum_{i=1}^n 1_{X_i\in B} \quad \mathop{\longrightarrow}_{n \to \infty}  \quad \mathbb{P}(Z\in B), \qquad \textrm{o√π } Z \sim \mathcal{N}(0,1) . \qquad \qquad \qquad (5)$$

> Nous avons vu que la cha√Æne de Markov √©tait stationnaire (les v.a. $X_i$ ont m√™me loi) et de loi stationnaire $\pi_\star$.  C'est cette loi stationnaire qui appara√Æt comme loi limite dans le th√©or√®me ergodique.

Concr√®tement, pour $n\gg 1$, l'histogramme de $X_1, \ldots, X_n$ est proche de la densit√© de $\mathcal{N}(0, 1)$. 

<span style="color:blue">Question 3.1.</span> Illustrer la proximit√© entre l'histogramme de l'√©chantillon $(X_i)_{1 \le i \le n}$ et la loi $\mathcal{N}(0,1)$.

#### B. Th√©or√®me ergodique et m√©thode de splitting ####


Si l'on modifie la dynamique du processus $X$ dans (3) en conditionnant √† l'appartenance √† un ensemble mesurable $A\subset \mathbb{R}$,  la convergence (5) reste valable (au sens conditionnel). Soit $A\subset \mathbb{R}$  un ensemble mesurable de mesure de Lebesgue strictement positive. 

Soit $\tilde X_0$ une v.a. de  loi gaussienne centr√©e r√©duite conditionn√©e √† appartenir √† $A$. Etant donn√©es une suite de v.a. $(Y_i)_i$ i.i.d. de loi $\mathcal{N}(0,1)$ et ind√©pendantes de $\tilde X_0$, on d√©finit pour pour tout $i \geq 1$,
$$
 \tilde X_{i+1/2}:=  \rho \tilde X_{i-1} +\sqrt{1-\rho^2} Y_i, \qquad Y_i \sim \mathcal{N}(0,1), 
$$
puis 
\begin{equation*}\label{eqAR(1)2} \tilde X_{i+1} := \begin{cases} \tilde X_{i+1/2} & \textrm{ si $\tilde X_{i+1/2}\in A$,}\\ \tilde X_{i} &\textrm{ sinon,}\end{cases} \qquad \qquad (6)
\end{equation*}
On a  
- pour tout $i \geq 0$, $\tilde X_i$ suit la loi gaussienne centr√©e r√©duite conditionn√©e √† appartenir √† $A$,  
- le th√©or√®me ergodique reste valable: pour tout $B\subset \mathbb{R}$  mesurable,  presque s√ªrement
\begin{equation*}\label{eqAR(1)TE2}
\frac{1}{n}\sum_{i=1}^n 1_{\tilde X_i \in B}  \quad \mathop{\longrightarrow}_{n \rightarrow \infty} \quad \mathbb{P}(Z\in B \,|\, Z\in A)\qquad \qquad   Z\sim \mathcal{N}(0, 1) \qquad (7)\end{equation*}


<span style="color:blue">Question 3.2.</span>  Prendre $A := [a,+\infty[$.
Illustrer la proximit√© entre l'histogramme de l'√©chantillon $( \tilde X_i)_{1 \le i \le n}$ et la loi conditionnelle de $Z$ sachant $\{Z \in A\}$, o√π $Z \sim \mathcal{N}(0,1)$. 

```python
## longueur de la chaine X
n = int(5e4)

## Param√®tres de l'auto-regressif 
rho = 0.7
rho_bar = np.sqrt(1-rho**2)

## Seuil pour la loi conditionnelle cible: Loi de X|X>a pour
a = .8

## Stockage des valeurs de la chaine
X = np.zeros(n)

## Initialisation de la chaine 
x_0 = np.random.randn(1)
while x_0 < a:
    x_0 = np.random.randn(1)
X[0] = x_0


## Simulation de la chaine par l'algorithme (6) 
Y = np.random.randn(n)
for i in range(1, n):
    newX = rho*X[i-1] + rho_bar*Y[i]
    X[i] = newX*(newX>=a) + X[i-1]*(newX<a)


## Affichage de l'histogramme 
# D√©finition du nombre de b√¢tons 
MinX, MaxX = min(X), max(X)
L = MaxX - MinX
sigma = np.sqrt(np.var(X))

N = int(round(n**(1./3.)*L/(3.49*sigma)))


# Loi cible : densit√© Gaussienne conditionnelle
x = np.linspace(MinX, MaxX, 100)
f_x_cond = sps.norm.pdf(x) / sps.norm.sf(a)
plt.plot(x, f_x_cond, "r", linewidth=1.5, label="Densite conditionnelle")
plt.axvline(x=a, color = "k", label="a")

# histogramme
plt.hist(X, bins=N, density=True, histtype='step',label="Histogramme des $X_i$")

#On choisit la position de la legende
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2, fancybox=True, shadow=True)
plt.show()

```

<span style="color:blue">M√©thode de splitting.</span> Pour calculer $\mathbb{P}(Z\in A)$, l'id√©e du _splitting_ est  d'utiliser une suite d√©croissante de "cibles"
$\mathbb{R}=A_0\supset A_1\supset\cdots\supset A_k=A$
et de calculer $\mathbb{P}(Z\in A)$ avec la formule 

\begin{equation*} 
\mathbb{P}(Z\in A)=\prod_{\ell=1}^k \mathbb{P}(Z\in A_{\ell}\,|\, Z\in A_{\ell-1}) \qquad \qquad (8)
\end{equation*}

o√π chaque probabilit√© conditionnelle est approch√©e via la limite (7), √† l'aide de v.a. $(\tilde X_i^{(\ell)})_i$  simul√©es selon l'algorithme (6) appliqu√© avec $A \leftarrow A_{\ell-1}$.

<span style="color:blue">Question 3.3.</span>  Illustrer la m√©thode de splitting pour le calcul de $\mathbb{P}(Z>a)$ pour $Z\sim \mathcal{N}(0,1)$, $a=4$ et $\rho = 0.6$.

La valeur de cette probabilit√© est connue et accessible via les fonctions `scipy.stats.norm.pdf` ou `scipy.stats.norm.sf`. On souhaite ici tester les performances d'un algorithme d'estimation bas√© sur la repr√©sentation (8) et sur le r√©sultat de convergence (7), dans un cas simple.

Plus pr√©cis√©ment, on pourra tester l'algorithme suivant :  
- Etape 1: on id√©ntifie des ensembles $A_l$ raisonnables. En utilisant un nombre $n$ de tirages pas tr√®s √©lev√© ($n=500$ ou $n=1000$ par exemple), on estime, en se basant sur la convergence (7), des valeurs
$$
-\infty=a_0<a_1<\cdots <a_k < a \leq a_{k+1}
$$
telles que pour tout $\ell$,
$$
\mathbb{P}(Z>a_\ell\,|\,Z>a_{\ell-1})
\simeq 0.1.
$$

> Indication : penser au fait que le calcul de $a_\ell$ (une fois que $a_{\ell-1}$ est connu) revient √† l'estimation d'un quantile! Pour cela, on pourra admettre que les th√©or√®mes limites de l'Exercice 1 restent vrais pour l'√©chantillon $(\tilde X_i^{(\ell)})_i$.


- Etape2: on estime alors chacune des probabilit√©s conditionnelles $\mathbb{P}(Z>a_\ell\,|\,Z>a_{\ell-1})$ via une m√©thode encore fond√©e sur (7), en utilisant maintenant un nombre de tirages plus √©lev√© (par exemple de l'ordre de $n=5.10^4$). On estime aussi, pour la derni√®re probabilit√© : $\mathbb{P}(Z>a \,|\, Z > a_k)$. 

> Attention, en toute rigueur, $k$ est un nombre al√©atoire; ce que les notations utilis√©es dans cette section ne refl√®tent pas, pour simplifier l'expos√©.*


3. On en d√©duit  $\mathbb{P}(Z>a)$ via la formule (8), et on compare le r√©sultat obtenu √† la valeur approch√©e obtenue par un Monte-Carlo simple r√©alis√© avec autant de simulations (c√†d avec $n \, k$ simulations), ainsi qu'√† la valeur exacte donn√©e par la fonction `scipy.stats.norm.sf`.

```python
###################
# Etape 1
###################

def NiveauxSplitting(a, seuil, rho, n):
    """
    Fonction qui renvoie une estimation des niveaux
    de splitting a_1, a_2, ..., a_k definis dans la Q3.3
    
    Ces niveaux sont des quantiles d'une loi conditionnelle.
    
    M√©thode: on utilise l'inversion de la fonction de repartition empirique de 
    cette loi (comme fait dans la question 1.4 de la partie 1) afin d'estimer un
    quantile par le quantile empirique.
    """
    ## Estimation du premier niveau a_1: c'est le 
    ## quantile d'une loi Gaussienne non conditionnelle.
    rho_bar = np.sqrt(1-rho**2)
    X = np.random.randn(1)
    Y = np.random.randn(n)
    for i in range(n):
        X = np.append(X, rho*X[-1] + rho_bar*Y[i])    
    X.sort()
    
    quantiles = np.array([])    
    quantiles = np.append(quantiles, X[int(np.ceil((1-seuil)*n))-1] )
    print("niveau 1 : {}".format(quantiles[-1]))
    
    ## Estimation des autres niveaux 
    print("Inside while")
    while quantiles[-1] < a:
        ## Pour initialiser la chaine courante:
        ## on recupere la premiere valeur de la chaine
        ## precedente qui depasse le quantile
        print("plus gde valeur de la chaine {}".format(max(X)))
        startingPoint = X[X > quantiles[-1]][0]
        X = np.array([startingPoint])
        
        ## Simulation de la chaine avec rejet
        Y = np.random.randn(n)
        for i in range(n):
            newX = rho*X[-1] + rho_bar*Y[i]
            X = np.append(X, newX*(newX>quantiles[-1]) + X[-1]*(newX<=quantiles[-1]))

        ## On calcule le niveau suivant comme le quantile (1-seuil)
        X.sort()
        quantiles = np.append(quantiles, X[int(np.ceil((1-seuil)*n))-1])
        print("niveau suivant : {}".format(quantiles[-1]))
        
    ## On selectionne les niveaux a_1, a_2,... strictement en dessous de a
    quantiles = quantiles[:-1]
    print("\n valeurs successives des niveaux")
    print(quantiles)
    ## On rajoute a
    niveaux = np.append(quantiles, a)
    
    return niveaux



if __name__ == "__main__":    
    ## Parametres du probleme. ATTENTION : si a grand et rho petit (ex. rho=0.1) pbme sur le calcul des derniers niveaux car la chaine n'accepte quasiment plus
    a = 5
    rho = 0.6
    rho_bar =  np.sqrt(1-rho**2)
    ## Choix de la progression de recherche des seuils
    seuil = 0.1
    
    print("On cherche les niveaux de splitting approximatifs... \n")
        
    n = int(1e4) 
    niveaux = NiveauxSplitting(a, seuil, rho, n)
        
    print("\n On a trouv√© les niveaux de splitting \n")
    print("Nombre de niveaux = {} : ".format(len(niveaux)-1))
    
    ## Une fois les niveaux a_1, ..., a_k trouv√©s, on lance
    ## la simulation des chaines pour estimer chaque probabilit√©
    ## conditionnelle, avec une valeur de n plus importante
    n = int(515)
    Y = np.random.randn(n)
    X = np.random.randn(1)
    
    ## Niveau 0: pas de rejet
    for i in range(n):
        X = np.append(X, rho*X[-1] + rho_bar*Y[i])
        
    P = [ np.mean(X > niveaux[0]) ]
    
    ## Niveau l
    k = len(niveaux)
    
    for l in range(1, k):
        a_l_moins_1 = niveaux[l-1]
        
        ## On recupere la premiere valeur de la chaine precedente 
        ## qui depasse le niveau l-1:
        ## pour initialiser la chaine au niveau courant
        startingPoint = X[X > a_l_moins_1][0]
        
        X = np.array([startingPoint])
        Y = np.random.randn(n)
    
        for i in range(n):
            newX = rho*X[-1] + rho_bar*Y[i]
            X = np.append(X, newX*(newX > a_l_moins_1) + X[-1]*(newX <= a_l_moins_1))
        
        P.append( np.mean(X > niveaux[l]) )
    
    # estimateur de P(X > a) par methode de splitting
    print("-"*40)
    print("\n Valeur des probabilites successives")
    print(P)
    Proba_emp_split = np.prod(P)
    
    # estimateur Monte-Carlo avec k*n tirages
    Proba_emp_MonteCarlo = np.mean( np.random.randn(k*n)>a )
    
    # valeur exacte
    p = sps.norm.sf(a)
    
    print("-"*40)
    
    print("Exact probability = {:1.3e}".format(p))
    
    print("Splitting method gives = {:1.3e}".format(Proba_emp_split))
    
    print("Simple Monte-Carlo gives = {:1.3e}".format(Proba_emp_MonteCarlo))
    
```

## Exercice 4. Estimation de la densit√©
***

Dans cet exercice, on consid√®re une suite $(X_i)_{i\ge 1}$ de al√©atoires i.i.d. dont on suppose qu'elles ont une densit√© $f$ sur $\R$. Afin d'estimer $f$, pour chaque valeur de $n$, on va construire, √† partir des observations $X_1, \ldots, X_n$, une densit√© $f_n$.
Cette densit√© pourra, selon la strat√©gie choisie, √™tre construite via diff√©rents proc√©d√©s, et on verra dans quelle mesure, pour $x\in \R$,
$$
f_n(x)\underset{n\to\infty}{\longrightarrow} f(x).
$$

#### A. Histogramme ####

Pour chaque valeur de $n$, on fixe une subdivision $a_{n,0}<a_{n,1}<\cdots$ de $\R$, et $f_n$ est la fonction constante sur chaque intervalle $]a_{n,k-1}, a_{n,k}]$ telle que  pour tout $k$,

$$\int^{a_{n,k}}_{a_{n,k-1}}f_n(x)\textrm{d} x
=
\frac 1 n 
\textrm{Card}\{i\in \{1,¬†\ldots, n\}: a_{n,k-1} < X_i\le a_{n,k}\}
= F_n(a_{n,k}) - F_n(a_{n,k-1}).
$$

#### B. Fen√™tres glissantes ####
Pour chaque valeur de $n$, on fixe $h_n>0$ (la _largeur de fen√™tre_) et on d√©finit la densit√© $f_n$ par la formule :
 
$$\label{def_fenetre_gli}\forall x\in \R,¬†\qquad f_n(x) := 
\frac 1 {n \, h_n}
\textrm{Card}\Big\{i\in \{1,¬†\ldots, n\}: x-\frac{h_n}{2}< X_i\le x+\frac{h_n}{2}\Big\}. \qquad \qquad (9)
 $$

#### C. Comment choisir la largeur des colonnes de l'histogramme ou des fen√™tres glissantes ? ####
